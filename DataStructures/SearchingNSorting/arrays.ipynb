## Linear search

```python
def linear_search(arr, target):
    size = len(arr)
    for index in range(0, size):
        if (arr[index] == target):
            return index
        
    return -1
```

```python
result = linear_search([12,31,34,19,10,11], 10)
print(result)
```

## Binary search

```python
def binary_search(sorted_arr, target):
    start = 0
    end = len(sorted_arr) - 1
    
    while (start <= end):
        mid = (end + start)//2
        if (sorted_arr[mid] == target):
            return mid
        elif (sorted_arr[mid] > target):
            end = mid-1
        elif (sorted_arr[mid] < target):
            start = mid + 1

    return -1
```

```python
arr_2 = [12,412,345,24,5,4,34]
sorted_arr = sorted(arr_2)
result = binary_search(sorted_arr, 345)
```
