```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./shapes.jpg")
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray = cv2.Canny(image_gray, 80,150)
```

```python
retval, binary = cv2.threshold(image_gray, 200,255, cv2.THRESH_BINARY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
countours , hireachy = cv2.findContours(gray, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# For bounding boxes using contours
sorted_contours = sorted(countours, key=cv2.contourArea, reverse=True)
# (image, mode, method)
# mode -  hireachy relationship between contours
# method - how countours points are stored
```

```python
print(len(sorted_contours))
```

```python
contours = sorted_contours[1:10]
```

```python
contour_image = image.copy()

for contour  in contours:
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(contour_image, (x,y),(x+w,y+h),(0,0,255),2)
```

```python
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))




























```
