```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image =  cv2.imread("./input.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
```

```python
horziontal = np.array([[-1,-1,-1],
                       [-15 , -15, -15],
                       [1,1,1]])

image_filtered = cv2.filter2D(image, -1, horziontal)
```

```python
plt.imshow(horziontal, cmap='gray')
```

## BLURRING

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
kernel_size = 15
```

```python
blurred_image = cv2.blur(image, (kernel_size, kernel_size))
```

```python
plt.imshow(blurred_image, cmap='gray')
```

```python
# Gaussian blur
gaussian_blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0) # Image, kernelsize, sigma max
```

```python
plt.imshow(gaussian_blurred_image, cmap='gray')
```
