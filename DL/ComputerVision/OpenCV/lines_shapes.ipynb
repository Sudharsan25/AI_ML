```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
height, width = 500, 700
blue = (255,127,0)
canvas = np.full((height, width,3),blue, dtype=np.uint8)
```

```python
cv2.line(canvas, (50,0), (650,650),(255,255,255),5)
```

```python
canvas_rgb = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)
plt.imshow(canvas_rgb)
```

```python
cv2.rectangle(canvas_rgb, (100,100),(300,300), (0,255,255),3) # -1 to fil the rectangle
```

```python
plt.imshow(canvas_rgb)
```

```python
cv2.ellipse(canvas, (350,350), (100,50),45, 0,360, (0,255,0),3)
plt.imshow(canvas)
```

```python
points = np.array([[150,400],[300,400],[255,300]],np.int32)
```

```python
reshaped = points.reshape((-1,1,2))
```

```python
cv2.polylines(canvas, [reshaped],True, (255,0,0),3)
```

```python
plt.imshow(canvas)
```

```python
text = "Shapes"
font = cv2.FONT_HERSHEY_COMPLEX

# bottom left corner of text 
org = (50,100)
font_scale = 1.5
color = (255,255,255)
thickness = 2
```

```python
cv2.putText(canvas_rgb,text,org,font,font_scale,color,thickness,cv2.LINE_4)
```

```python
plt.imshow(canvas_rgb)
```
