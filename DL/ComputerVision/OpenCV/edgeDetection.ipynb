```python
import cv2
import matplotlib.pyplot  as plt
```

## EDGE DETECTION USING SOBEL, LAPLACIAN, CANNY EDGE DETECTOR

```python
image = cv2.imread("./input.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3) # 1, 0 for X axis deravative
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3) # 0, 1 for Y axis deravative
```

```python
sobel_filtered = cv2.magnitude(sobel_x, sobel_y)
```

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=3)
laplacian_abs = cv2.convertScaleAbs(laplacian)
```

```python
canny_edges = cv2.Canny(image, 100, 200) # Threshold 1 and Threshold 2
```

```python
titles = ['Original', 'Sobel','Laplcian', 'Canny']
images = [image, sobel_filtered, laplacian_abs,canny_edges]
```

```python
plt.figure(figsize=(10,8))
for i in range (len(titles)):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
