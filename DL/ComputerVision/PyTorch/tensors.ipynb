```python
import torch

# 0D tensor / Scalar

a = 2

scalar = torch.tensor(42)
print("Scalar tensor", scalar)
print("Scalar Dimenson", scalar.ndim)
```

```python
# 1 D tensor / Vectors

vector = torch.tensor([1,2,3])
print("Vector tensor", vector)
print("Vector Dimenson", vector.ndim)
print("Vector Shape", vector.shape)
```

```python
# 2 D tensor /  Matirx

matrix = torch.tensor([[1,2],[3,4]])
print("matrix tensor", matrix)
print("matrix Dimenson", matrix.ndim)
print("matrix Shape", matrix.shape)
```

```python
# GPU ACCeleration

if torch.cuda.is_available():
    gpu_tensors = vector.to('cuda')
    print("Tensors on GPU:", gpu_tensors)
else:
    print("CUDA is not avalaibale on this device")
```

```python
# Tensor properties

print("Data type of matrix", matrix.dtype)
print("Device of tensor", gpu_tensors.device)
print("Size", matrix.size())
print("No.of elements:", matrix.numel())
```

```python
# Accessing elements in a tensor

#Indexing
# tensor[row, column]

tensor = torch.tensor([[1,2,3],[4,5,6],[3,4,5]])
print(tensor.ndim)
print(tensor.shape)
```

```python
print("Element at (0,1)",tensor[0,1]) # First Element
print("First row:", tensor[0]) # First row
print("First column:", tensor[:,0]) # First column
```

```python
# Slicing
# tensor[rowStart:rowEnd, columnStart:columnEnd]

print("First two rows",tensor[0:2]) # or tensor[:2]
print("First two columns", tensor[:, :2]) 
print("Middle elements:",tensor[1,1:3])
```

```python
# Complex indexing
# Get the (0,1) and (2,2)

print("Get the elements at (0,1) and (2,2)",tensor[[0,2],[1,2]])
```

```python
# Boolean indexing

mask = tensor > 2
print(mask)
```

```python
# Updating tensor values

print("Before:", tensor)
tensor[0,1] = 25
print("After:", tensor)
```

```python
# torch.index_select

indices = torch.tensor([0,2])
rows = torch.index_select(tensor, dim=0, index=(indices))
print(rows)
```

```python
print(tensor[:,::2]) # Every second element
```

```python
# Generate random integers

random_ints = torch.randint(0,10,(5,5)) # randint(low, high, size) # High is exclusive

print(random_ints)
print(random_ints.shape)
print(random_ints.ndim)
```

```python
# Generate random numbers from a normal distribution

normal_rand = torch.normal(0.5, 2, (3,3)) # normal(mean, std, size)

print(normal_rand)
print(normal_rand.shape)
print(normal_rand.ndim)
```
