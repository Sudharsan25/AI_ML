```python
# Accessing elements in a tensor

#Indexing
# tensor[row, column]

tensor = torch.tensor([[1,2,3],[4,5,6],[3,4,5]])
print(tensor.ndim)
print(tensor.shape)
```

```python
print("Element at (0,1)",tensor[0,1]) # First Element
print("First row:", tensor[0]) # First row
print("First column:", tensor[:,0]) # First column
```

```python
# Slicing
# tensor[rowStart:rowEnd, columnStart:columnEnd]

print("First two rows",tensor[0:2]) # or tensor[:2]
print("First two columns", tensor[:, :2]) 
print("Middle elements:",tensor[1,1:3])
```

```python
# Complex indexing
# Get the (0,1) and (2,2)

print("Get the elements at (0,1) and (2,2)",tensor[[0,2],[1,2]])
```

```python
# Boolean indexing

mask = tensor > 2
print(mask)
```

```python
# Updating tensor values

print("Before:", tensor)
tensor[0,1] = 25
print("After:", tensor)
```

```python
# torch.index_select

indices = torch.tensor([0,2])
rows = torch.index_select(tensor, dim=0, index=(indices))
print(rows)
```

### Tensor Manipulation

* Reshaping
* Slicing
* Joining and Spliting
* Transposing and permuting dimension

```python
import torch
```

```python
# Reshaping tensors

original_tensor = torch.tensor([1,2,3,3,4,3,34,4,4])
print(original_tensor.ndim)
```

```python
reshaped_tensor = original_tensor.reshape(3,3)
print(reshaped_tensor.ndim)  # Reshape to 3x3
```

```python
reshaped_tensor
```

```python
# view 

reshaped_tensor_view = original_tensor.view(3,3) # original_tesnor must be contiguous in memory (contiguous means the tensor is stored in a single block of memory)
print(reshaped_tensor_view.ndim)  # View to 3x3
print(reshaped_tensor_view) 
## reshape and view are similar but view is a view of the original tensor and does not create a new tensor
## Whereas reshape creates a new tensor with the same data but different shape
```

```python
original_tensor.is_contiguous()
```

```python
# Slicing

sliced_tensor = reshaped_tensor[:,:1]
print("Sliced tensor:", sliced_tensor)
```

```python
# Joining tensors

# torch.cat merges tesnors along an exisiting dimension

tensor_1 = torch.tensor([[1,1],[0,0]])
tensor_2 = torch.tensor([[1,0],[0,1]])
```

```python
cat_rows = torch.cat((tensor_1, tensor_2), dim=0)
cat_columns = torch.cat((tensor_1, tensor_2), dim=1)

print("Concatenated rows:\n", cat_rows)
print("Concatenated columns:\n", cat_columns)

# Dimension does not change
print("Shape of concatenated rows:", cat_rows.shape)
print("Shape of concatenated columns:", cat_columns.shape)

print("Dimension of concatenated rows:", cat_rows.ndim)
print("Dimension of concatenated columns:", cat_columns.ndim)
```

```python
# Stack

# torch.stack merges tensors along a new dimension, which increases the number of dimensions by 1

stack_rows = torch.stack((tensor_1, tensor_2), dim=0)
stack_columns = torch.stack((tensor_1, tensor_2), dim=1)

print("Concatenated rows:\n", stack_rows)
print("Concatenated columns:\n", stack_columns)

# Dimension does not change
print("Shape of concatenated rows:", stack_rows.shape)
print("Shape of concatenated columns:", stack_columns.shape)

print("Dimension of concatenated rows:", stack_rows.ndim)
print("Dimension of concatenated columns:", stack_columns.ndim)
```

```python
# Splitting tensors
# torch.chunk spits tensor into equal parts along a specified dimension
# torch.split splits tensor into specified number of parts along a specified dimension which can be unequal

original_tensor = torch.arange(12)

chunks = torch.chunk(original_tensor, 5)  # Split into 3 equal parts
print("Chunks:", chunks)
```

```python
split_tensor = torch.split(original_tensor, 5)  # Split into parts of size 4
```

```python
print("Split tensor:", split_tensor)
```

```python
# transpose and permuting tensors

# transpose swaps the dimensions of a tensor (m x n) to (n x m)
# permute rearranges the dimensions of a tensor in a specified order

original_tensor = torch.arange(24).reshape(12,2)

transpose = original_tensor.transpose(0,1)

print(original_tensor)
print(transpose)
```

```python
permute = original_tensor.permute(1,0) 
```

```python
# cloning and detaching tensors


original_tensor = torch.tensor([[1,2],[3,4]])
cloned_tensor = original_tensor.clone()  # Creates a new tensor with the same data
detached_tensor = original_tensor.detach()  # Creates a new tensor that shares the same data

print("Original tensor:", original_tensor)
print("Cloned tensor:", cloned_tensor) 
```
