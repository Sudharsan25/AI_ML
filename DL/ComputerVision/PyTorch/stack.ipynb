```python
import torch
```

```python
tensor_1 = torch.tensor([1,2,3])
tensor_2 = torch.tensor([4,5,6])
```

```python
print(tensor_1, tensor_2)
```

```python
vstack = torch.stack((tensor_1, tensor_2), dim=0)
```

```python
hstack = torch.stack((tensor_1, tensor_2), dim=1)
```

```python
print(vstack)
```

```python
print(hstack) 
print(tensor_1.ndim)
print(hstack.ndim)# adds an extra dimension to the output tensor
```

```python
# stacking 2d tensors

tensor2d_1 = torch.tensor([[1,2],[3,4]])
tensor2d_2 = torch.tensor([[5,6],[7,8]])
```

```python
vstack2d = torch.stack((tensor2d_1, tensor2d_2), dim=0)
```

```python
hstack2d = torch.stack((tensor2d_1, tensor2d_2), dim=1)
```

```python
print(vstack2d)
print(tensor2d_1.ndim)
print(vstack2d.ndim)# adds an extra dimension to the output tensor
```

```python
print(hstack2d) # dim =1 of first tensor and dim=1 of second tensor are stacked into first row of the stacked tensor
```

```python
# concat

concatenated = torch.cat((tensor2d_1, tensor2d_2), dim=0) 
```

```python
print(concatenated)
print(tensor2d_1.ndim)
print(concatenated.ndim)# does not add an extra dimension to the stacked tensor
```
