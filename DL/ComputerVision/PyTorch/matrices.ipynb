### Matrix aggregation

* Obtaining meaningful stats like mean, std, and sum of matrix 
* Helps in reducing dimensions of tensors

```python
import torch
```

```python
matrix = torch.tensor([[1.,2.,3.],
                      [1.,2.,3.],
                      [1.,2.,3.]])
```

```python
print(matrix)
```

```python
matrix.sum()
```

```python
matrix.mean()
```

```python
matrix.median()
```

```python
matrix.max()
```

```python
# aggregation across specific dimensions
# Column -> dim = 1 (across rows)
# row -> dim = 0 (across columns)

matrix.sum(dim=1)
```

```python
matrix.sum(dim=0)
```

```python
matrix.cumsum(dim=1) # cumlative
```

```python
# Advance Aggregation

matrix[matrix>1].sum()
```

```python
matrix.nonzero() # Gives the list of index of the non-zero elements in the matrix
```

```python
matrix.shape
```

```python
max_val = matrix.max()
min_val = matrix.min()


normalized_matrix = (matrix - min_val) / (max_val - min_val)
print(normalized_matrix) 
```
