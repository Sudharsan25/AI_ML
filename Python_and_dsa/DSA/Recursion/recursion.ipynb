```python
def printNumFive(num):
    print(num)
    printNumFour(num-1)

def printNumFour(num):
    print(num)
    printNumThree(num-1)

def printNumThree(num):
    print(num)
    printNumTwo(num-1)

def printNumTwo(num):
    print(num)
    printNumOne(num-1)

def printNumOne(num):
    print(num)
```

```python
printNumFive(5) # This is achieved by chaining function call in another function (Basic concept of recursion )
```

```python
## using recursion

def printNum(num):
    print(num)
    if (num==1):
        return
    
    printNum(num-1)


printNum(5)
```

## Factorial of a no.

```python
def factorial(n):
    if (n==1): ## base case
        return 1

    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
## sum of N numbers using recursion


def sumOfNusingRecursion(n):
    if(n==1):
        return 1
    
    return n + sumOfNusingRecursion(n-1)
```

```python
print(sumOfNusingRecursion(10))
```
