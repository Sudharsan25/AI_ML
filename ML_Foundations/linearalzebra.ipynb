```python
# linear alzebra
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) # linspace allows us to create points between 0 and 40 with 1000 points
```

```python
# linear equations 
# d = 2.5t
# d = 3.5 (t-5)t
```

```python
d_r = 2.5 * t
```

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('Distance vs Time')
plt.xlabel('Time (mins)')
plt.ylabel('Distance (km)')
ax.set_xlim(0, 40)
ax.set_ylim(0, 120)
ax.plot(t, d_r, label='Distance of R', color='blue')
ax.plot(t, d_s, label='Distance of S', color='red')
plt.axvline(x=30, color='green', linestyle='--', label='Meeting Point')
plt.axhline(y=75, color='purple', linestyle='--', label='Distance of S at Meeting Point')
plt.show()
```
